```python
import cv2 as cv
import numpy as np

def connected_components_stats_demo(src):
    cv.imshow('source image',src)
    src = cv.GaussianBlur(src, (3, 3), 0)#进行高斯去噪
    gray = cv.cvtColor(src, cv.COLOR_BGR2GRAY)#转换成灰度图
    ret, binary = cv.threshold(gray, 0, 255, cv.THRESH_BINARY | cv.THRESH_OTSU)#使用大津算法进行自动阈值寻找并二值化灰度图
    print('ret',ret)#显示阈值
    cv.imshow("binary", binary)#显示生成的二值图像
    #统计二值联通组件的信息
    num_labels, labels, stats, centers = cv.connectedComponentsWithStats(binary, connectivity=8, ltype=cv.CV_32S)
    
    colors = []#初始化颜色
    for i in range(num_labels):
        b = np.random.randint(0, 256)
        g = np.random.randint(0, 256)
        r = np.random.randint(0, 256)
        colors.append((b, g, r))
    colors[0] = (0, 0, 0)
    
    image = np.copy(src)
    for t in range(1, num_labels, 1):
        x, y, w, h, area = stats[t]
        cx, cy = centers[t]
        cv.circle(image, (np.int32(cx), np.int32(cy)), 2, (0, 255, 0), 2, 8, 0)#画出中心点
        cv.rectangle(image, (x, y), (x+w, y+h), colors[t], 1, 8, 0)#画出每个连通区域的矩形边框
        cv.putText(image, "num:" + str(t), (x, y), cv.FONT_HERSHEY_SIMPLEX, .5, (0, 0, 255), 1)#给连通区域添加文字说明
        print("label index %d,x:%d,y:%d,w:%d,h:%d, area : %d"%(t, x,y,w,h,area))#打印出每个连通区域的信息

    cv.imshow("colored labels", image)
    cv.imwrite('D:\data of rice.png',image)
    print("total rice : ", num_labels - 1)

file=r'C:\Users\luo yong chen\Pictures\rice.png'
input=cv.imread(file)
connected_components_stats_demo(input)
cv.waitKey(0)
cv.destroyAllWindows()
```
