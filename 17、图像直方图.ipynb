```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt


def custom_hist(gray):
    h, w = gray.shape
    hist = np.zeros([256], dtype=np.int32)
    print(hist)
    for row in range(h):      #遍历整个图像的像素
        for col in range(w):
            pv = gray[row, col]
            hist[pv] += 1       #相同大小的像素值加一

    y_pos = np.arange(0, 256, 1, dtype=np.int32)
    plt.bar(y_pos, hist, align='center', color='r', alpha=0.5)#align是条形的位置，有中心和边缘两个位置
    plt.xticks(y_pos, y_pos)#设置名字
    plt.ylabel('Frequency')
    plt.title('Histogram')

    # plt.plot(hist, color='r')
    # plt.xlim([0, 256])
    plt.show()


def image_hist(image):
    cv.imshow("input", image)
    color = ('blue', 'green', 'red')
    for i, color in enumerate(color):
        hist = cv.calcHist([image], [i], None, [256], [0, 256])#参数：image必须用方括号括起来，i是直方图的通道数，第三个参数是mask
                                    #第四个参数表示直方图分成多少份，第五个参数是像素值的范围
        plt.plot(hist, color=color)
        plt.xlim([0, 256])
    plt.show()


filename=r'C:\Users\luo yong chen\Pictures\tree.jpg'
src=cv.imread(filename)
gray = cv.cvtColor(src, cv.COLOR_BGR2GRAY)
print('灰度图像的像素个数：',gray.shape)
cv.imshow("input", gray)
custom_hist(gray)
image_hist(src)
cv.waitKey(0)
cv.destroyAllWindows()
```
